統計今年各月出貨重量

```python
import pandas as pd

def TRAN_DATE(days) :
    DATE_LIST = days.split("/")
    if len(DATE_LIST[1]) < 2 :
        DATE_LIST[1] = "0" + DATE_LIST[1]
    return "{}/{}".format(DATE_LIST[0], DATE_LIST[1])

data = pd.read_csv("0800Q.csv")
data["Order_weight(MT)"] = data["訂單重量(KG)"].map(lambda x : float(x.replace(",",""))*0.001)
data = data[~data["客戶訂單號碼"].str.contains("庫存單", na=False)]
CHART_SOURCE = data[["Order_weight(MT)", "訂單交期"]].copy()
CHART_SOURCE["Order_Monthly"] = CHART_SOURCE["訂單交期"].apply(TRAN_DATE)
FINAL_TABLE = CHART_SOURCE[["Order_weight(MT)", "Order_Monthly"]].groupby("Order_Monthly").sum()
FINAL_TABLE
```

```python
print("TOTAL_SHIPPING_WEIGHT : ",CHART_SOURCE["Order_weight(MT)"].sum())
```

以下為統計期間各公司訂單總金額(暫不使用)

```python
data.rename(columns={data.columns[3]:"Company"}, inplace = True)
INCOME_SOURCE = data[["下單客戶", "Company", "訂單金額", "幣別"]].copy()
INCOME_SOURCE["訂單金額"] = INCOME_SOURCE["訂單金額"].map(lambda x : float(x.replace(",","")))
END_INCOME = INCOME_SOURCE.groupby("Company").sum()
END_INCOME["幣別"] = END_INCOME["幣別"].map(lambda x : x[:3] if len(x) > 3 else x )
END_INCOME["下單客戶"] = END_INCOME["下單客戶"].map(lambda x : x[:4] if len(x) > 3 else x )
```

```python
NUM = []
for i in range(len(END_INCOME["幣別"])) : 
    if END_INCOME["幣別"][i] == "USD":
        NUM.append(int(END_INCOME["訂單金額"][i]*30))
    elif END_INCOME["幣別"][i] == "EUR" :
        NUM.append(int(END_INCOME["訂單金額"][i]*32.5))
    elif END_INCOME["幣別"][i] == "SEK" :
        NUM.append(int(END_INCOME["訂單金額"][i]*2.8))
    else :
        NUM.append(END_INCOME["訂單金額"][i])
END_INCOME["TOTAL_VALUE(NTD)"] = NUM
END_INCOME.sort_values(by = ["TOTAL_VALUE(NTD)"], ascending=False)
```

```python
print("今年度目前營業額 : ", END_INCOME["TOTAL_VALUE(NTD)"].sum())
```

年度接單與去年度相比

```python
from sqlalchemy import create_engine

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', "December"]

def Verify_Order_Name():
    oracle_connection_string = (
        "oracle+cx_oracle://spselect:select@192.168.1.242:1526/?service_name=sperpdb")
    engine = create_engine(oracle_connection_string)
    query = "SELECT SC_NO, CST_REFE_NO FROM V_SCH0200Q_ORD"
    df = pd.read_sql_query(query, engine.connect())
    df.columns = [col.upper() for col in df.columns]
    return df
```

```python
import pandas as pd


raw_order = pd.read_excel("0710Q_2025.xlsx")

ORDER_NAME_df = Verify_Order_Name()
Mapping_order = dict(zip(ORDER_NAME_df["SC_NO"], ORDER_NAME_df["CST_REFE_NO"]))

first_filter_order = raw_order[["客戶代號", "公司S/C",  "訂單日期", "確認日期","接單數量(M)", "接單重量(KGS)", "接單金額"]]
first_filter_order.loc[:, "訂單日期"] = pd.to_datetime(first_filter_order["訂單日期"])
first_filter_order.loc[:, "確認日期"] = pd.to_datetime(first_filter_order["確認日期"])
first_filter_order.loc[:, "Month"] = first_filter_order.loc[:, "確認日期"].dt.month_name()
first_filter_order.loc[:, "公司S/C"] = first_filter_order["公司S/C"].astype(str).str.strip()
first_filter_order["Order_name"] = first_filter_order["公司S/C"].map(Mapping_order)
first_filter_order = first_filter_order[~first_filter_order["Order_name"].str.contains("庫存單", na=False)]
final = first_filter_order.copy()

# according to weight
summary = pd.pivot_table(
    final,
    index = '客戶代號',          # rows
    columns = 'Month',           # columns
    values = '接單重量(KGS)',     # values to sum
    aggfunc = 'sum',             # aggregation function
    fill_value=0               # fill missing combinations with 0
)

# according to cash
summary_Cash = pd.pivot_table(
    final,
    index = '客戶代號',          # rows
    columns = 'Month',           # columns
    values = '接單金額',     # values to sum
    aggfunc = 'sum',             # aggregation function
    fill_value = 0               # fill missing combinations with 0
)

summary = summary.reindex(columns=month_order)
summary_Cash = summary_Cash.reindex(columns=month_order)

# 1. Add "年度總結" column (row sum)
summary['年度總結'] = summary.sum(axis=1)
summary['年度總結(%)'] = (summary['年度總結'] / summary['年度總結'].sum()) * 100
summary['年度總結(%)'] = summary['年度總結(%)'].round(2)

summary_Cash ['年度總結'] = summary_Cash .sum(axis=1)
summary_Cash ['年度總結(%)'] = (summary_Cash ['年度總結'] / summary_Cash ['年度總結'].sum()) * 100
summary_Cash ['年度總結(%)'] = summary_Cash ['年度總結(%)'].round(2)
# 2. Sort by "年度總結" descending
summary = summary.sort_values('年度總結', ascending=False)
summary_Cash = summary_Cash.sort_values('年度總結', ascending=False)
```

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

raw_order_2024 = pd.read_excel("0710Q_2024.xlsx")
raw_order_2025 = pd.read_excel("0710Q_2025.xlsx")

request_column = ["客戶代號", "公司S/C",  "訂單日期", "確認日期","接單數量(M)", "接單重量(KGS)", "接單金額"]
filter_order_2024 = raw_order_2024[request_column]
filter_order_2025 = raw_order_2025[request_column]
filter_order_2025.loc[:, "公司S/C"] = filter_order_2025["公司S/C"].astype(str).str.strip()
filter_order_2025.loc[:, "Order_name"] = filter_order_2025["公司S/C"].map(Mapping_order)
filter_order_2025 = filter_order_2025[~filter_order_2025["Order_name"].str.contains("庫存單", na=False)]



filter_order_2024.loc[:, "確認日期"] = pd.to_datetime(filter_order_2024["確認日期"])
filter_order_2025.loc[:, "確認日期"] = pd.to_datetime(filter_order_2025["確認日期"])

filter_order_2024.loc[:, "Month"] = filter_order_2024["確認日期"].dt.month_name()
filter_order_2025.loc[:, "Month"] = filter_order_2025["確認日期"].dt.month_name()

final_2024 = filter_order_2024.copy()
final_2025 = filter_order_2025.copy()

# according to weight
summary_2024 = pd.pivot_table(
    final_2024 ,
    index = '客戶代號',          # rows
    columns = 'Month',           # columns
    values = '接單重量(KGS)',     # values to sum
    aggfunc = 'sum',             # aggregation function
    fill_value=0               # fill missing combinations with 0
)
summary_2025 = pd.pivot_table(
    final_2025 ,
    index = '客戶代號',          # rows
    columns = 'Month',           # columns
    values = '接單重量(KGS)',     # values to sum
    aggfunc = 'sum',             # aggregation function
    fill_value=0               # fill missing combinations with 0
)

merged = pd.merge(
    summary_2024,
    summary_2025,
    left_index = True,
    right_index = True,
    suffixes=('_2024', '_2025'),
    
    how = 'outer'   # In case some customers only appear in one year
).fillna(0)
merged.to_excel("merged.xlsx")
growth = pd.DataFrame(index = merged.index)

for month in month_order:
    col_2024 = f"{month}_2024"
    col_2025 = f"{month}_2025"

    if col_2024 in merged.columns and col_2025 in merged.columns: 
        growth[month] = (merged[col_2025] - merged[col_2024]) / merged[col_2024]
        growth[month] = growth[month].replace(float('inf'), -2).fillna(0) # no order last year
    else:
        growth[month] = 0   # If not available, just fill 0

growth = growth.round(2)  # Round and turn into percentage directly
#growth = growth.map(lambda x: f"{x:.1f}%" if pd.notnull(x) else 'N/A')
growth = growth.map(lambda x : "No order last year" if x == -2 else x)
Z_store_path = r"Z:\業務部\業務一課\H-訂單\3. 接單統計\2025年度\年度接單重量對照表(2024x2025).xlsx"

growth.to_excel(Z_store_path)

with pd.ExcelWriter(Z_store_path, engine="xlsxwriter") as writer:
    summary.to_excel(writer, sheet_name = "Summary_重量(庫存單已扣)")
    summary_Cash.to_excel(writer, sheet_name = "Summary_金額(庫存單已扣)")
    growth.to_excel(writer, sheet_name = '2025VS2024重量比較')
    workbook = writer.book
    worksheet = writer.sheets['2025VS2024重量比較']
    percent_format = workbook.add_format({'num_format': '0%'})
    worksheet.set_column("A1:M20", None, percent_format)
    # Conditional format: green background for > 0
    worksheet.conditional_format("B2:M20", {
        'type':     'cell',
        'criteria': '>',
        'value':    0,
        'format':   workbook.add_format({'bg_color': '#C6EFCE', 'font_color': '#006100'})
    })

    # Conditional format: red background for < 0
    worksheet.conditional_format("B2:M20", {
        'type':     'cell',
        'criteria': '<',
        'value':    0,
        'format':   workbook.add_format({'bg_color': '#FFC7CE', 'font_color': '#9C0006'})
    })
    worksheet.conditional_format("B2:M20", {
        'type':     'text',
        'criteria': 'equal to',
        'value':    "No order last year",
        'format':   workbook.add_format({'bg_color': '#BFBFBF', 'font_color': '#FFFFFF'})
    })


```
