```python
PACKAGE_WEIGHT_DF = df[["PMT_NO", "PCK_NAME","PCK_LEVEL", "CTN_WT"]]
PACKAGE_WEIGHT_DF
```

```python
order_df = pd.read_sql_query("SELECT * FROM ssl_cst_orde_d where SC_NO = '25010001'", connection)
CAL_WEIGHT_DF = order_df[["CST_PART_NO", # item code
                          "PMT_NO",   # CTN code
                          "SML_PMT_NO", # inner box code
                          "SMALL_PACK_QTY", # quantity per inner box
                          "KEGS", # quanityt of outer box or carotn
                          "BOXES", # quantity of inner box
                          "PLT_QTY", # pallet quantity
                          "ORDER_WEIG" # screw weight 
                         ]].copy().fillna(0)
CAL_WEIGHT_DF
```

```python
CAL_WEIGHT_DF["PMT_NO_WEI"] = CAL_WEIGHT_DF["PMT_NO"].map(PACKAGE_WEIGHT_DF.set_index("PMT_NO")["CTN_WT"])
CAL_WEIGHT_DF["SML_PMT_NO_WEI"] = CAL_WEIGHT_DF["SML_PMT_NO"].map(PACKAGE_WEIGHT_DF.set_index("PMT_NO")["CTN_WT"])
CAL_WEIGHT_DF = CAL_WEIGHT_DF.fillna(0)
```

```python
CAL_WEIGHT_DF["PCK_SUM_WEI"] = (CAL_WEIGHT_DF["BOXES"] * CAL_WEIGHT_DF["SML_PMT_NO_WEI"]) + \
                               (CAL_WEIGHT_DF["KEGS"] * CAL_WEIGHT_DF["PMT_NO_WEI"])

CAL_WEIGHT_DF[["CST_PART_NO", "KEGS", "PMT_NO_WEI", "BOXES", "SML_PMT_NO_WEI", "PCK_SUM_WEI"]]
```

```python
ORDER_WEI_SUM = CAL_WEIGHT_DF["PCK_SUM_WEI"].sum() + CAL_WEIGHT_DF["PLT_QTY"].sum()*20 + \
                CAL_WEIGHT_DF["ORDER_WEIG"].sum()
```

```python
SCREW_WEI = CAL_WEIGHT_DF["ORDER_WEIG"].sum()
```

```python
DOUBLE_PALLETS = CAL_WEIGHT_DF.loc[CAL_WEIGHT_DF["BOXES"] == 0, "PLT_QTY"].sum()
TRIPLE_PALLETS = CAL_WEIGHT_DF.loc[CAL_WEIGHT_DF["BOXES"] > 1, "PLT_QTY"].sum()

DOUBLE_PALLETS, TRIPLE_PALLETS
```
