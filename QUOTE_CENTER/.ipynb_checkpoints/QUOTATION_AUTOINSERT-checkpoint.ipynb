```python
import sqlite3
# Connect to the SQLite database (creates a new database if it doesn't exist)
DATABASE_ADDRESS = "QUOTATION_DATABASE.db"
conn = sqlite3.connect(DATABASE_ADDRESS)

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Create a table called "test"
cursor.execute('''
    CREATE TABLE IF NOT EXISTS 
    CUSTOMER_PRODUCT_SUMMARY (
        ID INTEGER PRIMARY KEY AUTOINCREMENT,
        PRODUCT_CODE TEXT,
        DRAWING_NUMBER TEXT,
        SIZE TEXT,
        RECESS_PARTIAL_THREAD TEXT,
        MOQ INTEGER,
        QUANTITY REAL, WEIGHT REAL,
        M_BOX REAL, KG_BOX REAL,
        BOX_TYPE TEXT, BOX_PRICE REAL,
        WIRE_DIAMETER REAL, WIRE_PRICE REAL,
        FORMING_PRICE REAL, FORMING_PIECE_WEIGHT REAL, FORMING_TOTAL REAL, FORMING_GROSS REAL,
        CUT_SLUTTING_LOSS REAL, CUT_SLUTTING_PRICE_M REAL, CUT_SLUTTING_PRICE_KG REAL, CUT_SLUTTING_GROSS REAL,
        HEAT_TREATMENT_LOSS REAL, HEAT_TREATMENT_PRICE REAL, HEAT_TREATMENT_GROSS REAL,
        PLATING_LOSS REAL, PLATING_PRICE_KG REAL, PLATING_GROSS REAL,
        PACKAGE_LOSS REAL, PACKAGE_PRICE REAL, PACKAGE_GROSS REAL,
        TRANSACTION_CONDITION TEXT, OPERATION_FEE REAL , FIRST_GROSS REAL,
        ROLLER_SORTING_LOSS REAL, ROLLER_SORTING_PRICE REAL, ROLLER_SORTING_GROSS_KG REAL, ROLLER_SORTING_GROSS_M REAL, 
        OPTICAL_SORTING_LOSS REAL, OPTICAL_SORTING_PRICE REAL, OPTICAL_SORTING_GROSS_M REAL,
        INTEREST_RATE REAL, INTEREST_GROSS REAL, 
        COLLATED_LOSS REAL, COLLATED_PRICE, COLLATED_GROSS,
        PROFIT_RATE REAL, PROFIT_GROSS REAL,
        EXCHANGE_RATE REAL, EXCHANGE_RATE_GROSS REAL, REBATE REAL, TOTAL_PRICE_M REAL, 
        DESCRIPTION TEXT,
        ORDER_ADDRESS TEXT, QUOTE_DATE TEXT, ORDER_TYPE TEXT,
        AGENT TEXT, CUSTOMER_CODE TEXT, ORDER_NUMBER TEXT, 
        SERIAL_NUMBER REAL)
    ''')
conn.commit()
conn.close()
```

```python
from openpyxl import Workbook
import sqlite3

def SET_QUOTE_FORMULA(PANDAS, OUTPUTFILE) :
    wb = Workbook()
    ws = wb.active
    # write the column into excel sheet
    for r_index, row in enumerate(PANDAS.columns, start = 1):
        ws.cell(row = 1, column = r_index, value = PANDAS.columns[r_index - 1])
    # write the dataframe into excel sheet
    for r_idx, row in enumerate(PANDAS.itertuples(), start = 2): 
        for c_idx, value in enumerate(row[1:], start = 1):  
            ws.cell(row = r_idx, column = c_idx, value=value)
    # write the formula into excel sheet
    for r_index in range(PANDAS.shape[0]) :
        row_n = r_index + 2
        ws.cell(row = row_n, column = 7, value = f"=F{row_n}*O{row_n}")
        ws.cell(row = row_n, column = 9, value = f"=O{row_n}*H{row_n}")
        ws.cell(row = row_n, column = 16, value = f"=N{row_n}/O{row_n}")
        ws.cell(row = row_n, column = 17, value = f"=ROUNDUP((P{row_n} + M{row_n}), 2)")
        ws.cell(row = row_n, column = 20, value = f"=ROUNDUP(S{row_n}/O{row_n}, 2)")
        ws.cell(row = row_n, column = 21, value = f"=ROUNDUP(Q{row_n}*(1 + R{row_n}) + T{row_n}, 2)")
        ws.cell(row = row_n, column = 24, value = f"=ROUNDUP(U{row_n}*(1 + V{row_n}) + W{row_n}, 2)")
        ws.cell(row = row_n, column = 27, value = f"=ROUNDUP(X{row_n}*(1 + Y{row_n}) + Z{row_n}, 2)")
        ws.cell(row = row_n, column = 29, value = f"=ROUNDUP(K{row_n}/I{row_n}, 2)")
        ws.cell(row = row_n, column = 30, value = f"=ROUNDUP(AA{row_n}*(1 + AB{row_n}) + AC{row_n}, 2)")
        ws.cell(row = row_n, column = 33, value = f"=ROUNDUP((AD{row_n} + AE{row_n} + AF{row_n}), 2)")
        ws.cell(row = row_n, column = 36, value = f"=ROUNDUP(AG{row_n}*(1 + AH{row_n}) + AI{row_n}, 2)")
        ws.cell(row = row_n, column = 37, value = f"=ROUNDUP(AJ{row_n}*O{row_n}, 2)")
        ws.cell(row = row_n, column = 40, value = f"=ROUNDUP(AK{row_n}*(1 + AL{row_n}) + AM{row_n}, 2)")
        ws.cell(row = row_n, column = 42, value = f"=ROUNDUP(AN{row_n}*(1 + AO{row_n}), 2)")
        ws.cell(row = row_n, column = 44, value = f"=ROUNDUP(AP{row_n}*(1 + AQ{row_n}), 2)")
        ws.cell(row = row_n, column = 47, value = f"=ROUNDUP(AR{row_n}*(1 + AS{row_n}) + AT{row_n}, 2)")
        ws.cell(row = row_n, column = 49, value = f"=AU{row_n}/AV{row_n}")
        ws.cell(row = row_n, column = 51, value = f"=ROUNDUP(AW{row_n}*(1 + AX{row_n}), 2)")
    wb.save(OUTPUTFILE)

    
    
```

```python
import pandas as pd 


def SEARCH_ITEM(ITEMLIST, database, quotation_list): #(詢價品項excel, 連結資料庫位置, 輸出column)
    RESULT = pd.DataFrame()
    with  sqlite3.connect(database) as connection :
        cursor = connection.cursor()
        MOTHER = pd.read_excel(ITEMLIST)
        start = 2
        for productcode in MOTHER["客戶代號"]:
            qury = (f"SELECT {', '.join(quotation_list)} FROM CUSTOMER_PRODUCT_SUMMARY WHERE PRODUCT_CODE = ?")
            item = pd.read_sql_query(qury, connection, params=(productcode,))
            
            if item.shape[0] == 0:
                item = pd.DataFrame([[0] * len( RESULT.columns)], columns = RESULT.columns)
                item.iloc[0, 0] = productcode
                RESULT = pd.concat([RESULT, item])
            elif item.shape[0] == 1 :
                RESULT = pd.concat([RESULT, item])
            else :
                RESULT = pd.concat([RESULT, item])
            start += 1
        
        SET_QUOTE_FORMULA(RESULT, "output_file.xlsx")
k = SEARCH_ITEM("quote_item.xlsx", DATABASE_ADDRESS, QUOTATION_COLUMN)
        
```
